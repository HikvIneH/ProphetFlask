```python
import pandas as pd
import numpy as np
from fbprophet import Prophet
import datetime
import matplotlib.pyplot as plt

import os.path
import fix_yahoo_finance as yf
import pandas as pd

plt.rcParams['figure.figsize']=(20,10)
plt.style.use('ggplot')
```

```python
startDate = datetime.datetime(2010, 1, 4).date()
endDate = datetime.datetime.now().date()
startDate
```

```python
def get_historical_stock_price(stock):
    print ("Getting historical stock prices for stock ", stock)
    stockData = yf.download(stock, startDate, endDate)
    stockData.to_csv("./static/data/"+stock+'-'+str(startDate)+'to'+str(endDate)+".csv")
    return stockData
```

```python
stock = 'aapl'
if os.path.isfile("./static/data/"+stock+'-'+str(startDate)+'to'+str(endDate)+".csv") == False: 
    df_historical = get_historical_stock_price(stock)
else:
    df_historical = pd.read_csv("./static/data/"+stock+'-'+str(startDate)+'to'+str(endDate)+".csv")
```

```python
dfr = df_historical.round(3)
df = dfr.filter(['Close'])
dfa = dfr.filter(['Close'])
```

```python
df['ds'] = df.index
dfa['ds'] = df.index
dfa.tail()
```

```python
#log transform 'Close' from non-stationary to stationary
df['y'] = np.log(df['Close']).round(3)
df.tail()
```

```python
#model = Prophet(weekly_seasonality=True, daily_seasonality=True, yearly_seasonality=True)
model = Prophet()
model.fit(df)
```

```python
num_days = 365
```

```python
future = model.make_future_dataframe(periods=num_days)
forecast = model.predict(future)
    
print (forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())    
```
